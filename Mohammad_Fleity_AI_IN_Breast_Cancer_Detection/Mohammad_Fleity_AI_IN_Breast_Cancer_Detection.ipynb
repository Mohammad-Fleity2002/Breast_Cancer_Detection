```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=2)
print(x.shape,train_x.shape,test_x.shape)
```

<h3 style="color:green">Standarizing the data</h3>
<i style="color:blue">Standardization is a data preprocessing technique used to scale features to have a <b style="color:red">mean=0</b> and a <b style="color:red">standard deviation=1</b>. This helps improve the performance of machine learning algorithms that are sensitive to the scale of the features, such as <b style="color:red">SVMs</b>, <b style="color:red">logistic regression</b>, and <b style="color:red">neural networks</b></i>

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
train_x_std=scaler.fit_transform(train_x)
test_x_std=scaler.transform(test_x)
```

<h3 style="color:green"> Building the Neural Network</b>

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),# input layer
                        # Flatten the matrix of feature x values to an array so that each neuron represent a feature
                                             keras.layers.Dense(20,activation='relu'),# one hidden layer
                                             keras.layers.Dense(2,activation='sigmoid')# one output layer 2 neurons one malignant and one belignant 
                                             ])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# create the model 
```

<h2 style="color:green">Momentum:</h2> 

This algorithm is used to accelerate the gradient descent algorithm.. 

&emsp; <b style="color:red"><i>w<sub>t+1</sub>=w<sub>t</sub>−αm<sub>t</sub></i></b>

where,

&emsp; <b style="color:red"><i>m<sub>t</sub>=βm<sub>t-1</sub>+(1−β)[∂L/∂w<sub>t</sub>]</i></b>

<p style="background-color:#c5c5c5;padding:10px" >
<b>mt </b>= aggregate of gradients at time t [current] (initially, mt = 0)</br>
<b>mt-1 </b>= aggregate of gradients at time t-1 [previous]</br>
<b>Wt </b>= weights at time t</br>
<b>Wt+1 </b>= weights at time t+1</br>
<b>αt </b>= learning rate at time t</br> 
<b>∂L </b>= derivative of Loss Function</br>
<b>∂Wt</b> = derivative of weights at time t</br>
<b>β </b>= Moving average parameter (const, 0.9)
</p>

<div style="color:blue"><h2>Advantages:</h2>
<ul>
<li><b>Faster convergence: </b>faster convergence, especially in valleys or along flat surfaces.</li>
<li><b>Handles noise better.</b></li>
<li><b>Simple to implement.</b></li>
</ul>
</div>
<div style="color:red;"><h2>Disadvantages:</h2>
<ul>
<li><b>Manual learning rate tuning: </b> Requires careful selection of the learning rate to ensure convergence.</li>
<li><b>No adaptive learning rate:</b> It does not adapt the learning rate based on individual parameters, making it inefficient for problems with sparse gradients or varying scales</li>
<li><b>May overshoot minima:</b>High momentum can sometimes cause updates to overshoot the optimal solution.</li>
</ul>
</div>

<h2 style="color:green">Root Mean Square Propagation (RMSP): </h2>

Root mean square prop or RMSprop is an adaptive learning algorithm that tries to improve AdaGrad.

&emsp; <b style="color:red"><i>w<sub>t+1</sub>=w<sub>t</sub>-(&alpha;<sub>t</sub> / &#8730;(&nu;<sub>t</sub> + &epsilon;)) &times; [ ∂L / ∂w<sub>t</sub> ]</i></b>

where,

&emsp; <b style="color:red"><i>&nu;<sub>t</sub>=β&nu;<sub>t-1</sub>+(1−β)[∂L/∂w<sub>t</sub>]<sup>2</sup></i></b>

<p style="background-color:#c5c5c5;padding:10px" >
<b>&epsilon; </b>= A small positive constant (10-8)</br>
<b>&nu;<sub>t</sub> </b>= sum of square of past gradients.[i.e sum(∂L/∂Wt-1)] (initially, Vt = 0)</br>
<b>Wt </b>= weights at time t</br>
<b>Wt+1 </b>= weights at time t+1</br>
<b>αt </b>= learning rate at time t</br> 
<b>∂L </b>= derivative of Loss Function</br>
<b>∂Wt</b> = derivative of weights at time t</br>
<b>β </b>= Moving average parameter (const, 0.9)
</p>
<div style="color:blue"><h2>Advantages:</h2>
<ul>
<li><b>Adaptive learning rate: </b>Adjusts the learning rate for each parameter, making it effective for non-stationary problems.</li>
<li><b>Efficient for sparse gradients:</b>Performs well on problems with sparse data and gradients, like <b>RNNs</b> and <b>deep networks.</b></li>
<li><b>Reduces oscillations.</b></li>
</ul>
</div>
<div style="color:red;"><h2>Disadvantages:</h2>
<ul>
<li><b>Learning rate decay: </b>Learning rates decrease over time, which can cause slow convergence toward the end.</li>
<li><b>Hyperparameter tuning: </b>The decay rate (rho) and initial learning rate still require careful tuning.</li>
<li><b>May not generalize well: </b>It sometimes struggles with generalization compared to Adam.</li>
</ul>
</div>

<h2 style="color:green">Adam Optimizer:</h2> 
inherits the strengths or the positive attributes of the above two methods and builds upon them to give a more optimized gradient descent.

&emsp; <b style="color:red"><i>m<sub>t</sub>=β<sub>1</sub>m<sub>t-1</sub>+(1−β<sub>1</sub>)[∂L/∂w<sub>t</sub>]</i></b>

&emsp; <b style="color:red"><i>&nu;<sub>t</sub>=β<sub>2</sub>&nu;<sub>t-1</sub>+(1−β<sub>2</sub>)[∂L/∂w<sub>t</sub>]<sup>2</sup></i></b>

<p style="background-color:#c5c5c5;padding:10px" >
<b>1. &epsilon; </b>= A small positive constant to avoid 'division by 0' error when (vt -> 0). (10<sup>-8</sup>)</br>
<b>2. β<sub>1</sub> & β<sub>2</sub></b>= decay rates of average of gradients in the above two methods. (β<sub>1</sub> = 0.9 & β<sub>2</sub> = 0.999)</br>
<b>3. α </b>  — Step size parameter / learning rate (0.001)</br> 
</p>
<div style="color:blue"><h2>Advantages:</h2>
<ul>
<li><b>Combines Momentum and RMSProp:</b> Adam benefits from both momentum for faster convergence and adaptive learning rates for efficiency.</li>
<li><b>Faster convergence.</b></li>
<li><b>Default hyperparameters work well: </b><i>(beta1=0.9, beta2=0.999, and epsilon=1e-8).</i></li>
<li><b>Handles sparse gradients well:</b> Performs well on problems with sparse gradients <i>(e.g., NLP tasks).</i></li>
<li><b>Less hyperparameter tuning.</b></li>
</ul>
</div>
<div style="color:red;"><h2>Disadvantages:</h2>
<ul>
<li><b>Memory overhead:</b>  Adam requires storing additional moving averages for the first and second moments, increasing memory usage.</li>
<li><b>Non-stationary learning rates: </b>The learning rate decay in Adam can sometimes cause instability late in training.</li>
</ul>
</div>

```python
#Training the neural network
history=model.fit(train_x_std,train_y,validation_split=0.1,epochs=10)
```

<h2 style="color:green">Visualization</h2>

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc = 'lower right')
plt.title('model accuracy')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title('model loss ')
plt.legend(['ttraining data','validation data'],loc='upper right')
```

1. When you fit the StandardScaler using a <b>pandas DataFrame</b>, the scaler "remembers" the feature names from the columns of that DataFrame.
2. When you later call scaler.transform() with a <b>NumPy array</b>, the feature names are absent, which causes the warning.
